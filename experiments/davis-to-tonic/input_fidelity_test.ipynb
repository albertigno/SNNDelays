```python
import numpy as np
data = np.load('davis_frames_X.npy')
```

```python
data.shape
```

```python
import matplotlib.pyplot as plt

# Sum across the first dimension (axis=0), resulting in shape (1, 2, 32, 32)
summed = data.sum(axis=0)

# Plot each channel separately
fig, axes = plt.subplots(1, 2, figsize=(10, 5))
for i in range(2):
    axes[i].imshow(summed[0, i], cmap='gray')
    axes[i].set_title(f'Channel {i}')
    axes[i].axis('off')
plt.tight_layout()
plt.show()
```
